Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1
**


```
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative as derr
import scipy as sc
import sympy as sm
from sympy import *
```

```
def f(x):
  return x**5 + 4*sc.sin(2*x) + sc.cos(3*x+3)
x0 = 1
fd1 = derr(f, x0)
fd2 = derr(f, x0, n=2)
print(f'первая производная {fd1} \nвторая производная {fd2}\n')
```

(sin(2x+1))^5
на отрезке [-5; 5]

```
f = (sin(2*x+1))**5
plot(diff(f,x, 1), (x,-5,5), line_color='r', title='первая производная', ylim=(-3,3))
plot(diff(f,x, 2), (x,-5,5), line_color='g', title='вторая производная', ylim=(-25,25))
```
