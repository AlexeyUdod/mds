этап 2: взять два атрибута и посчитать скалярное произведение двух векторов (2 дома), начертить 2 дома как вектора в двухмерном пространстве, и сравниваем графически и через скалярное произведение.


```
import matplotlib.pyplot as plt
import numpy as np
```

```
x1 = boston.data[0][0:2]
x2 = boston.data[1][0:2]
print(x1.dot(x2))

data1 = boston.data[0:2]
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()
x = np.array(data1[:,0])
y = np.array(data1[:,1])
ax.plot([0,x[0]], [0,y[0]])
ax.plot([0,x[1]], [0,y[1]])
ax.scatter(x[0], y[0], c='b')
ax.scatter(x[1], y[1], c='r')
plt.show()
```

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')
x = np.array(data1[:,3])
y = np.array(data1[:,4])
z = np.array(data1[:,5])
ax.plot([0,x[0]], [0,y[0]], [0,z[0]])
ax.plot([0,x[1]], [0,y[1]], [0,z[1]])
ax.scatter(x[0], y[0], z[0], c='b', marker='o')
ax.scatter(x[1], y[1], z[1], c='r', marker='o')
plt.show()
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```
import numpy as np
import matplotlib.pyplot as plt

A = [[-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]]

T = [[1,0,0],
     [0,1,0],
     [300,200,1]]

A1 = np.array(A)
T = np.array(T)
A2 = A @ T

plt.plot(A2[:,0], A2[:,1], c='red')
plt.plot(A1[:,0], A1[:,1], c='green')
plt.legend(['tranformed', 'original'])
# plt.ylim([-200, 200])
# plt.xlim([-200, 200])
```

Уменьшить объект в два раза.

```
T = [[.5,0,0],
     [0,.5,0],
     [0,0,1]]

A1 = np.array(A)
T = np.array(T)
A2 = A @ T

plt.plot(A2[:,0], A2[:,1], c='red')
plt.plot(A1[:,0], A1[:,1], c='green')
plt.legend(['tranformed', 'original'])

plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
